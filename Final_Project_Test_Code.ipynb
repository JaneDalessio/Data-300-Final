## Necesssary Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Exploratory Data Analysis

Reading in the Seoul Bike Sharing Demand dataset from the UC Irvine Machine Learning Repository.

```python
seoul_bike_data = pd.read_csv("SeoulBikeData.csv", encoding = "latin-1")
```

### First Looks

Taking a look at the first 5 rows of the dataset.

```python
seoul_bike_data.head()
```

Taking a look for missing values.

```python
seoul_bike_data.info()
```

Getting some descriptive statistics for the dataset.

```python
descriptive_stats = seoul_bike_data.describe()
summary = descriptive_stats.loc[["mean", "std", "25%", "50%", "75%", "min", "max"],:]
summary.T
```

### Data Distributions (Continuous Variables)

#### Rented Bike Count

Taking a look at the distribution of the `Rented Bike Count` variable.

```python
sns.histplot(data = seoul_bike_data, x = "Rented Bike Count")
plt.title("Distribution of Rented Bike Count Data")
plt.show()
```

#### Temperature

Taking a look at the distribution of the `Temperature(°C)` variable.

```python
bins = np.arange(-20, 40, 1)
sns.histplot(data = seoul_bike_data, x = "Temperature(°C)", bins = bins)
plt.title("Distribution of Temperature Data")
plt.show()
```

#### Humidity

Taking a look at the distribution of the `Humidity(%)` variable.

```python
bins = np.arange(0, 100, 2.5)
sns.histplot(data = seoul_bike_data, x = "Humidity(%)", bins = bins)
plt.title("Distribution of Humidity Data")
plt.show()
```

#### Wind Speed
